# Data Cleaning for Fraud Detection Dataset

In this notebook, we will perform various data cleaning steps on the fraud detection dataset. This includes handling missing values, duplicates, and irrelevant columns.

---

## Step 1: Handling Missing Values

### Objective:
We need to check for any missing values in the dataset and decide on how to handle them.

### Approach:
- We check for missing values using the `isnull()` method combined with `sum()` to see the number of missing values in each column.
- We then drop rows with missing values using `df_cleaned.dropna()` if the missing values are minimal and not important for the analysis.


```python
df = pd.read_csv('/home/hoda/fintech_de_terraform/fintech-dashboard-terraform/data/creditcard.csv')
```

```python
df_cleaned = df.dropna()
```

```python
duplicates = df_cleaned[df_cleaned.duplicated(keep=False)]  # keep=False to mark all duplicates
print(f"Number of duplicate rows: {duplicates.shape[0]}")
```

```python
duplicates_sorted = duplicates.sort_values(by=['Class', 'Time'])
```

```python
print(duplicates_sorted)
```

```python
# Keep duplicates for fraudulent transactions (Class == 1), 
# but remove duplicates for non-fraudulent ones (Class == 0)
df_no_duplicates = df_cleaned[~(df_cleaned.duplicated(subset=['Time', 'Amount'], keep=False) & (df_cleaned['Class'] == 0))]
```

```python
# Check how many duplicates are left in the cleaned data
print(f"Number of remaining duplicates: {df_no_duplicates.duplicated().sum()}")
```

```python
# Count the number of fraudulent transactions (Class == 1)
fraudulent_transactions = df_cleaned[df_cleaned['Class'] == 1]
print(f"Number of fraudulent transactions: {fraudulent_transactions.shape[0]}")
```
